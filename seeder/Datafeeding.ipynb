```python
import pandas as pd
import mysql.connector
import numpy as np
```

```python
mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="sqlpassword"  
)

mycursor = mydb.cursor()
mycursor.execute("use messManagement")
```

```python
dfs = pd.read_excel("Student.xlsx")
dfs
```

```python
table = np.array(dfs)
```

```python

for row in table:
    sentence = 'INSERT INTO Student VALUES ("' + row[0] + '", "' + row[1] + '", "' + row[2] + '","' + str(row[3]) + '", "' + row[4] + '");'
    mycursor.execute(sentence)
mycursor.execute("SELECT * FROM Student")
myresult = mycursor.fetchall()

for x in myresult:
  print(x)
```
